```python
# handle Date of Journey: Try to Seperate Date, Month and Year
# Feature ENG.
df['date']=df['Date_of_Journey'].str.split('/').str[0] ## Gives Date
df['month']=df['Date_of_Journey'].str.split('/').str[1] ## Gives Month
df['year']=df['Date_of_Journey'].str.split('/').str[2] ## Gives Year
```

```python
df.info()
```

```python
# Convert to Int
df['date']=df['date'].astype(int)
df['month']=df['month'].astype(int)
df['year']=df['year'].astype(int)
```

```python
# Drop Date of Journey column == NO Require
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']= df['Arrival_Time'].apply(lambda x:x.split(' ')[0]) ## TO Remove the Date from Arrival TIme
```

```python
## Transform Arrival_Time
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_Minute'] = df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
# Dep_Time
df['Dep_hour'] = df['Dep_Time'].str.split(":").str[0]
df['Dep_minute'] = df['Dep_Time'].str.split(":").str[1]
```

```python
df['Dep_hour']=df['Dep_hour'].astype(int)
df['Dep_minute']=df['Dep_minute'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```
