# Nominal / One hot encoding 


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Create a Dataframe
df = pd.DataFrame({
    'color':['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
# Create an isntance of OneHot encoder
encoder = OneHotEncoder()
```

```python
## Perform Fit and Transform
encoded = encoder.fit_transform(df[['color']]).toarray() # Alphabatic Order
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
#When get new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df], axis=1)
```

# Ordinal Encoding


```python
# # THis is Rakable
# # For Example :
# High School - 1
# College - 2
# Graduate - 3
# Post-Graduate - 4
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.DataFrame({
    'size':['small','medium','large','medium','large','small']
})
```

```python
df
```

```python
encoder = OrdinalEncoder(categories=[['small','medium','large']]) ## Ordinal Encoder has parameter called Category to give rank.
encoder.fit_transform(df[['size']])
```

```python
encoder.transform([['small']])
```

Target Guided Ordinal Encoding

```python
df = pd.DataFrame({
    'city':['Mumbai','Surat','Vadodara','Surat','Vadodara','Mumbai'],
    'price':[200,150,100,180,150,220]
})
df
```

```python
mean_price = df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```

```python
df
```
